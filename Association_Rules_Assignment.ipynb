```python
retail = pd.read_excel('Online retail.xlsx')
retail
```

```python
# Checking for missing values
missing_values = retail.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
# Removing duplicates
data_cleaned = retail.drop_duplicates()
print("Cleaned Data:")
print(data_cleaned)
```

```python
# Splitting each transaction into a list of items
transactions = data_cleaned.iloc[:, 0].apply(lambda x: x.split(','))
print("Transactions:")
print(transactions)
```

```python
# Converting data to appropriate format
# For association rule mining, we need a one-hot encoded format
# Create a new DataFrame with unique items as columns
unique_items = []
for transaction in transactions:
  for item in transaction:
    if item not in unique_items:
      unique_items.append(item)

print("Unique Items:")
print(unique_items)
```

```python
encoded_transactions = pd.DataFrame(columns=unique_items)
# One-hot encode the transactions
for index, transaction in enumerate(transactions):
  row = [0] * len(unique_items)
  for item in transaction:
    try:
      item_index = unique_items.index(item)
      row[item_index] = 1
    except ValueError:
      pass  # Ignore unknown items
  encoded_transactions.loc[index] = row

print("Encoded Transactions:")
print(encoded_transactions)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Applying Apriori algorithm
frequent_itemsets = apriori(encoded_transactions, min_support=0.01, use_colnames=True)
print("Frequent Itemsets:")
print(frequent_itemsets)

# Generating association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
print("Association Rules:")
print(rules)
```

```python
# Applying Apriori algorithm with adjusted threshold
frequent_itemsets = apriori(encoded_transactions, min_support=0.05, use_colnames=True)
print("Frequent Itemsets (with adjusted support):")
print(frequent_itemsets)

# Generating association rules with adjusted thresholds
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=2)
print("Association Rules (with adjusted confidence and lift):")
print(rules)
```

```python
# Applying Apriori algorithm
frequent_itemsets = apriori(encoded_transactions, min_support=0.005, use_colnames=True)
print("Frequent Itemsets:")
print(frequent_itemsets)

# Generating association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
print("\nAssociation Rules:")
print(rules)

# Sort rules by lift
rules_sorted = rules.sort_values(by=['lift'], ascending=False)
print("\nTop rules by lift:")
print(rules_sorted.head(10))

# Filter rules based on confidence
high_confidence_rules = rules[rules['confidence'] > 0.6]
print("\nRules with high confidence:")
print(high_confidence_rules)

# Analyze antecedents and consequents
for index, row in rules_sorted.iterrows():
  antecedents = list(row['antecedents'])
  consequents = list(row['consequents'])
  print(f"\nAntecedents: {antecedents}, Consequents: {consequents}, Lift: {row['lift']}, Confidence: {row['confidence']}")

# Identify interesting patterns
target_product = 'mineral water'
interesting_rules = rules[rules['consequents'].apply(lambda x: target_product in x)]
print(f"\nRules with {target_product} in consequent:")
print(interesting_rules)
```

```python
# Insights into customer purchasing behaviour:
print("\nInsights:")
print("Based on the discovered association rules, we can infer the following about customer purchasing behaviour:")

# Checking for high lift values
high_lift_rules = rules_sorted[rules_sorted['lift'] > 2]
if not high_lift_rules.empty:
  print("- There are strong relationships between certain products, indicated by high lift values.")
  print("- Customers who purchase products in the antecedents are more likely to purchase products in the consequents.")

# Checking for high confidence values
high_confidence_rules = rules_sorted[rules_sorted['confidence'] > 0.6]
if not high_confidence_rules.empty:
  print("- High confidence values suggest a strong association between the antecedent and consequent.")
  print("-  Customers who buy the antecedent products are very likely to buy the consequent products.")

# Analyzing specific product relationships
if not interesting_rules.empty:
  print(f"- Customers who purchase {target_product} are likely to also purchase the products in the antecedents of the rules.")

# General observations
print("- The association rules can be used to understand which products are frequently purchased together.")
print("- This information can be used to optimize product placement, design targeted promotions, and improve overall customer experience.")
print("- By analyzing the rules, we can identify potential product bundles or recommend complementary products to customers.")
```
